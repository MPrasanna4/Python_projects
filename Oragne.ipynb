```python
# Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

correlation = data['Size (cm)'].corr(data['Weight (g)'])
print(f'Correlation between Size and Weight: {correlation}')

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Size (cm)', y='Weight (g)', hue='Variety', s=100)
plt.title('Correlation between Size and Weight of Fruit')
plt.xlabel('Size (cm)')
plt.ylabel('Weight (g)')
plt.grid()
plt.show()
```

```python
#How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?
correlation = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print(f'Correlation between Sweetness and Acidity: {correlation}')

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Brix (Sweetness)', y='pH (Acidity)', hue='Variety', s=100)
plt.title('Correlation between Sweetness and Acidity of Fruit')
plt.xlabel('Sweetness (Brix)')
plt.ylabel('Acidity (pH)')
plt.grid()
plt.axhline(y=4.0, color='r', linestyle='--', label='Neutral pH')
plt.legend()
plt.show()
```

```python
# How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
correlation = data['Softness (1-5)'].corr(data['Ripeness (1-5)'])
print(f'Correlation between Softness and Ripeness: {correlation}')

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Softness (1-5)', y='Ripeness (1-5)', hue='Variety', s=100)
plt.title('Correlation between Softness and Ripeness of Fruit')
plt.xlabel('Softness (1-5)')
plt.ylabel('Ripeness (1-5)')
plt.xticks(range(1, 6))  # Set x-ticks for clarity
plt.yticks(range(1, 6))  # Set y-ticks for clarity
plt.grid()
plt.show()
```

```python
#Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?
data['Blemishes (Y/N)'] = data['Blemishes (Y/N)'].astype('category')

# Calculate average ripeness and softness by blemish status
average_scores = data.groupby('Blemishes (Y/N)').agg(
    Average_Softness=('Softness (1-5)', 'mean'),
    Average_Ripeness=('Ripeness (1-5)', 'mean')
).reset_index()

# Display results
print("Average Softness and Ripeness by Blemish Status:")
print(average_scores)

# Visualization
plt.figure(figsize=(12, 6))

# Bar plot for Average Softness by Blemish Status
plt.subplot(1, 2, 1)
sns.barplot(data=average_scores, x='Blemishes (Y/N)', y='Average_Softness', palette='viridis', legend='False')
plt.title('Average Softness by Blemish Status')
plt.xlabel('Blemishes (Y/N)')
plt.ylabel('Average Softness (1-5)')
plt.xticks(rotation=45)

# Bar plot for Average Ripeness by Blemish Status
plt.subplot(1, 2, 2)
sns.barplot(data=average_scores, x='Blemishes (Y/N)', y='Average_Ripeness', palette='magma')
plt.title('Average Ripeness by Blemish Status')
plt.xlabel('Blemishes (Y/N)')
plt.ylabel('Average Ripeness (1-5)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness?
average_values = data.groupby('Variety').mean().reset_index()

# Display results
print("Average Size, Weight, Sweetness, Acidity and Ripeness by Variety:")
print(average_values)

# Visualization
plt.figure(figsize=(14, 10))

# Bar plots for each attribute
attributes = ['Size (cm)', 'Weight (g)', 'Brix (Sweetness)', 'pH (Acidity)', 'Ripeness (1-5)']
for i in range(len(attributes)):
    plt.subplot(3, 2, i + 1)
    sns.barplot(data=average_values.sort_values(by=attributes[i]), x='Variety', y=attributes[i], palette='viridis')
    plt.title(f'Average {attributes[i]} by Variety')
    plt.xlabel('Fruit Variety')
    plt.ylabel(attributes[i])
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
