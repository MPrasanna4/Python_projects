```python
# Do private school students perform better in exams than public school students?
def analyze_exam_performance(data): # By using function
    private_avg = data[data['School_Type'] == 'Private']['Exam_Score'].mean()
    public_avg = data[data['School_Type'] == 'Public']['Exam_Score'].mean()
    
    if private_avg > public_avg:
        print("Based on the data, private school students tend to perform better on exams.")
    elif public_avg > private_avg:
        print("Based on the data, public school students tend to perform better on exams.") 
    else: 
        print("No significant difference observed between private and public school students.")
    print(f"Average exam score for private school students: {private_avg}")
    print(f"Average exam score for public school students: {public_avg}")
df = pd.DataFrame(data)

analyze_exam_performance(df) 
```
