```python
#What are the trends in total purchase amount over time (based on Purchase Date)? Are there any seasonal patterns?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data['Purchase Date'] = pd.to_datetime(data['Purchase Date'])

# Group by date and sum the purchase amounts
daily_totals = data.groupby('Purchase Date')['Total Purchase Amount'].sum().reset_index()

# Set the Purchase Date as the index for time series analysis
daily_totals.set_index('Purchase Date', inplace=True)

# Resample to get monthly totals
monthly_totals = daily_totals.resample('M').sum()

# Plotting the trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_totals, x=monthly_totals.index, y='Total Purchase Amount')
plt.title('Total Purchase Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Purchase Amount ($)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Which product categories generate the highest revenue and which are the most popular based on quantity purchased?
data['Total Purchase Amount'] = data['Quantity'] * data['Product Price']

# Group by category to get total revenue and quantity sold
category_summary = data.groupby('Product Category').agg(
    Total_Revenue=('Total Purchase Amount', 'sum'),
    Total_Quantity_Sold=('Quantity', 'sum')).reset_index()

# Find the category with the highest revenue and quantity sold
highest_revenue_category = category_summary.loc[category_summary['Total_Revenue'].idxmax()]
highest_quantity_category = category_summary.loc[category_summary['Total_Quantity_Sold'].idxmax()]

# Display results
print("Total Revenue by Category:")
print(category_summary)
print("\nCategory with Highest Revenue:")
print(highest_revenue_category)
print("\nCategory with Highest Quantity Sold:")
print(highest_quantity_category)

# Visualization
plt.figure(figsize=(12, 6))

# Bar plot for Total Revenue by Category
plt.subplot(1, 2, 1)
sns.barplot(data=category_summary, x='Product Category', y='Total_Revenue', palette='viridis')
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)

# Bar plot for Total Quantity Sold by Category
plt.subplot(1, 2, 2)
sns.barplot(data=category_summary, x='Product Category', y='Total_Quantity_Sold', palette='magma')
plt.title('Total Quantity Sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#What are the most popular payment methods, and how do they relate to the total purchase amount or frequency of purchases?
payment_summary = data.groupby('Payment Method').agg(
    Total_Revenue=('Total Purchase Amount', 'sum'),
    Frequency=('Total Purchase Amount', 'count')
).reset_index()

# Find the most popular payment methods based on frequency and total revenue
most_popular_payment_method = payment_summary.loc[payment_summary['Frequency'].idxmax()]
highest_revenue_payment_method = payment_summary.loc[payment_summary['Total_Revenue'].idxmax()]

# Display results
print("Total Revenue and Frequency by Payment Method:")
print(payment_summary)
print("\nMost Popular Payment Method (by Frequency):")
print(most_popular_payment_method)
print("\nPayment Method with Highest Revenue:")
print(highest_revenue_payment_method)

# Visualization
plt.figure(figsize=(12, 6))

# Bar plot for Total Revenue by Payment Method
plt.subplot(1, 2, 1)
sns.barplot(data=payment_summary, x='Payment Method', y='Total_Revenue', palette='viridis')
plt.title('Total Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)

# Bar plot for Frequency of Purchases by Payment Method
plt.subplot(1, 2, 2)
sns.barplot(data=payment_summary, x='Payment Method', y='Frequency', palette='magma')
plt.title('Frequency of Purchases by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Frequency of Purchases')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Who are the top customers by total purchase amount or frequency, based on Customer ID and Customer Name?

#Calculate total revenue and frequency of purchases by customer
customer_summary = data.groupby(['Customer ID', 'Customer Name']).agg(
    Total_Revenue=('Total Purchase Amount', 'sum'),
    Frequency=('Total Purchase Amount', 'count')
).reset_index()

# Find the top customers based on total revenue and frequency
top_revenue_customer = customer_summary.loc[customer_summary['Total_Revenue'].idxmax()]
top_frequency_customer = customer_summary.loc[customer_summary['Frequency'].idxmax()]

# Display results
print("Total Revenue and Frequency by Customer:")
print(customer_summary)
print("\nTop Customer by Total Revenue:")
print(top_revenue_customer)
print("\nTop Customer by Frequency:")
print(top_frequency_customer)

# Visualization
plt.figure(figsize=(12, 6))

# Bar plot for Total Revenue by Customer
plt.subplot(1, 2, 1)
sns.barplot(data=customer_summary.sort_values(by='Total_Revenue', ascending=False), 
            x='Customer Name', y='Total_Revenue', palette='viridis')
plt.title('Total Revenue by Customer')
plt.xlabel('Customer Name')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)

# Bar plot for Frequency of Purchases by Customer
plt.subplot(1, 2, 2)
sns.barplot(data=customer_summary.sort_values(by='Frequency', ascending=False), 
            x='Customer Name', y='Frequency', palette='magma')
plt.title('Frequency of Purchases by Customer')
plt.xlabel('Customer Name')
plt.ylabel('Frequency of Purchases')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
