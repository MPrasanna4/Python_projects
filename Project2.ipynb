```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("netflix_movies.csv")# importing the excel filesin csv
data
```

```python
null_counts = data.isnull().sum()
print(null_counts)
```

```python


# Count the occurrences of each genre
genre_counts = data['type'].value_counts()

# Find the genre with the most movies and TV shows
most_popular_genre = genre_counts.index[0]
count = genre_counts[0]

print("The most popular genre is:", most_popular_genre)
print("Number of movies and TV shows in this genre:", count)
```

```python
#What is the distribution of content types (e.g., Movies vs. TV Shows) in the dataset?
df = pd.read_csv("netflix_movies.csv") 
def analyze_content_distribution(df):
    
    content_counts = df['type'].value_counts()# calculate the each count of the coulumn 
    print("Content type distribution:")
    print(content_counts)
    
    # Visualize using a bar chart
    plt.figure(figsize=(8, 5))
    plt.bar(content_counts.index, content_counts.values)
    plt.xlabel("Content Type")
    plt.ylabel("Count")
    plt.title("Distribution of Movie vs. TV Show Content")
    plt.show()



    
analyze_content_distribution(df)
```

```python
#What are the most common ratings, and how are they distributed across different types of content?
df = pd.read_csv("netflix_movies.csv") 
content_ratings = df.groupby("type")["rating"].value_counts(normalize=True)
#counts the occurrences of each rating value and normalizes the counts (divides by the total count) to get percentages.

# Print the distribution
print("Rating Distribution by Content Type:" , content_ratings )


# Find most common rating overall
most_common_rating = df["rating"].mode()[0]
print(f"\nMost common rating overall: {most_common_rating}")
```
