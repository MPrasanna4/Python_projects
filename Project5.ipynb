```python
#) What is the average price of cars across different brands, and which brands have the highest and lowest average prices?

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp



average_prices = data.groupby('Brand')['Price'].mean().reset_index()

# Find the brand with the highest average price
highest_avg_price = average_prices.loc[average_prices['Price'].idxmax()]

# Find the brand with the lowest average price
lowest_avg_price = average_prices.loc[average_prices['Price'].idxmin()]

# Display results
print("Average Prices by Brand:")
print(average_prices)
print("\nBrand with Highest Average Price:")
print(highest_avg_price)
print("\nBrand with Lowest Average Price:")
print(lowest_avg_price)

```

```python
#How does the condition of a car (Status) affect its price?
average_price_by_condition = data.groupby('Status')['Price'].mean()

# Print the results
print("Average price by condition:")
print(average_price_by_condition)

# Visualization of the price distribution based on car condition
plt.figure(figsize=(8, 6))
data.boxplot(by='Status', column='Price', grid=False)
plt.title('Price Distribution by Car Condition')
plt.suptitle('')
plt.xlabel('Type')
plt.ylabel('Price')
plt.show()
```

```python
#
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


correlation = data['Kilometers'].corr(data['Price'])
print(f'Correlation between Mileage and Price: {correlation}')

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Kilometers', y='Price')
plt.title('Car Price vs Mileage')
plt.xlabel('Mileage (Kilometers)')
plt.ylabel('Price ($)')
plt.grid()
plt.show()
```

```python
#How does engine capacity (CC) relate to car price and mileage?
#Calculate correlation coefficients
correlation_price_engine = data['CC'].corr(data['Price'])
correlation_mileage_engine = data['CC'].corr(data['Kilometers'])

print(f'Correlation between Engine Capacity and Price: {correlation_price_engine}')
print(f'Correlation between Engine Capacity and Mileage: {correlation_mileage_engine}')

# Visualization
plt.figure(figsize=(14, 6))

# Scatter plot for Engine Capacity vs Price
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x='CC', y='Price', color='blue')
plt.title('Engine Capacity vs Price')
plt.xlabel('CC')
plt.ylabel('Price')
plt.grid()

# Scatter plot for Engine Capacity vs Mileage
plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x='CC', y='Kilometers', color='red')
plt.title('Engine Capacity vs Mileage')
plt.xlabel('Engine Capacity (CC)')
plt.ylabel('Kilometers')
plt.grid()

plt.tight_layout()
plt.show()
```
